### Fold 1

```python
file_path = "/home/kumail/Energy_AI/Honey Model/Models/fold_1_train.csv"
df_train = pd.read_csv(file_path) #,'Absolute Concentration', 'Retention Time'
df_train
```

```python
file_path = "/home/kumail/Energy_AI/Honey Model/Models/fold_1_test.csv"
df_test = pd.read_csv(file_path) #,'Absolute Concentration', 'Retention Time'
df_test
```

```python
Y_train = df_train['Target']
X_train = df_train.drop(["Target"], axis=1)
X_train
```

```python
Y_test = df_test['Target']
X_test = df_test.drop(["Target"], axis=1)
X_test
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(X_train)
test_scaled = scaler.transform(X_test)
train_scaled
```

```python
CB_model = CatBoostRegressor(
        iterations=214,
        depth=8,
        learning_rate=0.2340820169767359,
        l2_leaf_reg=6.225477291266719,
        verbose=False,
        random_seed=42
    )
CB_model.fit(train_scaled, Y_train)
CB_y_pred =CB_model.predict(test_scaled)
r2 = r2_score(Y_test, CB_y_pred)
rmse = np.sqrt(mean_squared_error(Y_test,CB_y_pred))
mae = mean_absolute_error(Y_test, CB_y_pred)
print(f"R²: {r2:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
```
